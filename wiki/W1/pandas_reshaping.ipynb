Pandas를 사용하는 목적은 무엇인가?
- 주로 테이블 형태의 데이터를 조작하고 분석하는데 유용하다.
오늘 배우고자 하는 것과 그 이유는?
1. 목적에 따라 *주어진 데이터를 원하는 형태로 가공*할 수 있는 기술을 익히는 것
2. RDB는 저장에 유리한데 이를 분석에 용이하도록 피벗 테이블로 변환하는 작업의 의미와 필요성을 이해하는 것
3. 판다스를 이용해 재구성할 수 있는 데이터 레이아웃의 형태와 각각의 쓰임
4. 다른 API는 목적이 명확하다면 직관적으로 API를 찾아 사용할 수 있을 것 같다.

피벗 테이블은 **특정 기준과 관점**에 따라 데이터를 **요약, 집계, 재구성**하는데 사용해 분석에 활용한다.

```python
import pandas as pd
import numpy as np
import matplotlib
import matplotlib.pyplot as plt

source = 'https://raw.githubusercontent.com/softeerbootcamp4th/data_engineering_course_materials/refs/heads/main/missions/W1/mtcars.csv'
df = pd.read_csv(source)

# 데이터셋의 컬럼 중 이름이 없는 컬럼을 찾아서 적합한 이름으로 변경하세요.
df.rename(columns={'Unnamed: 0':'model'}, inplace=True)
df
```

```python
pivot_df = df.pivot_table(index='cyl', columns='vs', values=['qsec'], aggfunc=[np.mean, 'count'])
pivot_df
```
